## Q4) Is there a relationship of run time of movies vs domestic gross, popularity and production budget?

```python
# Importing the important libraries that we shall use in this investigation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline
import matplotlib
sns.set_style({'grid.color': 'black', 'axes.facecolor': 'white', 'figure.facecolor': 'white',
               'axes.edgecolor': 'black','axes.labelcolor': 'white', 'xtick.color': 'black',
               'ytick.color': 'black'})
```

```python
runtime_df = pd.read_csv("Data/group_data.csv")
```

```python
runtime_df['release_date'] = pd.to_datetime(runtime_df['release_date'])  #runtime vs domestic_gross , runtime vs rating , runtime vs prod_cost
```

```python
#Let us see the mean runtime of 100 highest grossing movies
top100 = runtime_df.nlargest(100,'domestic_gross')
print(top100['runtime'].mean())
print(top100['runtime'].median())
```

```python
#Let us see the mean runtime of 100 lowest grossing movies
bot100 = runtime_df.loc[runtime_df['domestic_gross'] > 0]
bot100 = bot100.nsmallest(100,'domestic_gross')
print(bot100['runtime'].mean())
print(bot100['runtime'].median())
```

```python
#Visualizing relationship between runtime and domestic_gross for top 100 movies
plt.figure(figsize=(13, 8))
sns.barplot(x='runtime', y='domestic_gross_in_mill', data=top100)
plt.xticks(rotation=90)
plt.show()
```

```python
#Visualizing relationship between runtime and popularity for top 100 movies
plt.figure(figsize=(13, 8))
sns.barplot(x='runtime', y='popularity', data=top100)
plt.xticks(rotation=90)
plt.show()
```

```python
#Visualizing relationship between runtime and domestic_gross for bottom 100 movies
plt.figure(figsize=(13, 8))
sns.barplot(x='runtime', y='domestic_gross_in_mill', data=bot100)
plt.xticks(rotation=90)
plt.show()
```

## Now let us examine a relationship between runtime and production budget

```python
runtime_df2 = runtime_df.drop(columns=['id','release_date','movie','genres','tconst','popularity','year_released','release_day','release_month','production_budget','domestic_gross'])
```

```python
corrMatrix = runtime_df2.corr() 
plt.figure(figsize=(13, 8)) 
sns.heatmap(corrMatrix, annot=True)
```

# Conclusion 

It seems that the **highest grossing** movies average to be around **123 minutes** while the **lowest grossing movies** average around the **95 minute** mark. The **most popular movies** in the **top 100 movies** have a runtime of **149 minutes**. 
If we take into consideration the most popular movies like Titanic, Avatar, All Marvel movies, this is what we would expect. 
As per our numbers, people normally like longer movies.
Also as we can see in the heatmap that the correlation coefficient of runtime to production budget is positively correlated and is **0.31** which is moderately strong.
